```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
import time
import pandas as pd

# date: 2022/11/22
# data: company, date, topic, content, source

result = {}
```

```python
# 01. GOGORO

COMPANY = 'GOGORO'
WEBSITE = 'https://www.gogoro.com/tw/news/archive/'
NEWS_FROM_COMPANY = True
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))


###### Get links #######################################

def get_links_in_page(page):
    global driver, result
    print(page)
    driver.get(f"{WEBSITE}?page={page}")
    time.sleep(2)
    
    links = set([link.get_attribute('href') for link in driver.find_elements(By.XPATH, '//*[@class="article-item"]/a')])

    for link in list(links):
        result[link] = get_topic_and_content(link)

        
###### Get contents #######################################
        
def get_topic_and_content(link):
    global driver
    
    driver.get(link)
    time.sleep(1)
    
    return {
        'company': COMPANY , 
        'date': driver.find_element(By.XPATH, '//*/time').text, 
        'topic': driver.find_element(By.CLASS_NAME, 'single-article-title').text, 
        'content': driver.find_element(By.CLASS_NAME, 'single-article-body').text.replace('\n', ''), 
        'source': COMPANY if NEWS_FROM_COMPANY else ''   
    }

    
for page in range(31, 74):
    get_links_in_page(page)
```

```python
# 03. 宏佳騰

COMPANY = '宏佳騰aeonmotor'
WEBSITE = 'https://ai.aeonmotor.com.tw/pages/brand'
NEWS_FROM_COMPANY = True
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
# titles = []
# links = []
# sources = []
# dates = []

###### Get links #######################################

def get_links_in_page(page):
    global driver, result, titles, links, sources, dates
    driver.get(f"{WEBSITE}")
    time.sleep(2)
    
    titles = [title.text for title in driver.find_elements(By.XPATH, '//*[@class="Grid-item Grid-text-item"]//h2')][1:]
    links = [link.get_attribute('href') for link in driver.find_elements(By.XPATH, '//*[@class="Grid-item Grid-text-item"]//*[@class="Grid-item-content"]//a')]
    content_elements = driver.find_elements(By.XPATH, '//*[@class="Grid-item Grid-text-item"]//*[@class="Grid-item-content"]//p')
    contents = [content_elements[i].text.split('\n')[0] for i in range(2, len(content_elements)) if content_elements[i].text != '閱讀全文...' and '20' in content_elements[i].text]
    sources = [content.split('｜')[0] for content in contents]
    dates = [content.split('｜')[1] for content in contents]
    
    for i in range(77 ,len(titles)):
        print(links[i])
        print(i)
#         print(get_topic_and_content(titles[i], dates[i],sources[i], links[i]))
        result[links[i]] = get_topic_and_content(titles[i], dates[i],sources[i], links[i])
        print('OK')

        
###### Get contents #######################################
        
def get_topic_and_content(title, date,source, link):
    global driver
    
    driver.get(link)
    time.sleep(1)
    
    
    try:
        content_elements = driver.find_elements(By.XPATH, '//*/p')
    
    except:
        return {
                'company': COMPANY , 
                'date': date, 
                'topic': title, 
                'content': link, 
                'source': source 
            }
    
    else:
        return {
            'company': COMPANY , 
            'date': date, 
            'topic': title, 
            'content': ''.join([content.text.replace('\n', '') for content in content_elements if content != [" "]]), 
            'source': source  
        } 

get_links_in_page(1)

# for i in range(19 ,len(titles)):
#     print(i)
# #     print(get_topic_and_content(titles[i], dates[i],sources[i], links[i]))
#     result[links[i]] = get_topic_and_content(titles[i], dates[i],sources[i], links[i])
```
